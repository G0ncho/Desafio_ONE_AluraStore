### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
tienda2.head()
tienda3.head()
tienda4.head()
```

```python
print(tienda.columns)
```

#1. Análisis de facturación



```python
# Cálculo para la Tienda 1
facturacion_t1 = tienda['Precio'].sum()

# Cálculo para la Tienda 2
facturacion_t2 = tienda2['Precio'].sum()

# Cálculo para la Tienda 3
facturacion_t3 = tienda3['Precio'].sum()

# Cálculo para la Tienda 4
facturacion_t4 = tienda4['Precio'].sum()

# Mostrar resultados
print(f"Facturación Total Tienda 1: {facturacion_t1:,.2f}")
print(f"Facturación Total Tienda 2: {facturacion_t2:,.2f}")
print(f"Facturación Total Tienda 3: {facturacion_t3:,.2f}")
print(f"Facturación Total Tienda 4: {facturacion_t4:,.2f}")
```

##Gráfico de Barras: Comparación de Facturación Total (Ingresos)

```python
import matplotlib.pyplot as plt

# Definir los datos (utilizando los resultados obtenidos)
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
facturacion = [
    1150880400.00,
    1116343500.00,
    1098019600.00,
    1038375700.00
]

plt.figure(figsize=(10, 6))
# Coloreamos la Tienda 4 de diferente color para destacarla
colores = ['skyblue', 'skyblue', 'skyblue', 'red']
plt.bar(tiendas, facturacion, color=colores)

plt.title('Comparación de la Facturación Total por Tienda')
plt.ylabel('Facturación Total (Miles de Millones)')
plt.xlabel('Tienda')

# Formatear el eje Y para una mejor lectura de números grandes
plt.ticklabel_format(style='sci', axis='y', scilimits=(0, 0))
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

**La Tienda 4** es la que menos ingresos genera, lo que la convierte en la principal candidata para ser vendida y justificar la inversión del Sr. Juan.

# 2. Ventas por categoría

```python
# Analizar la cantidad de productos vendidos por Categoría en cada tienda
print("Top 5: Categorías Más Populares por Tienda \n")

# Tienda 1
print("--- Tienda 1 (Mayor Facturación) ---")
print(tienda['Categoría del Producto'].value_counts().head())
print("-" * 35)

# Tienda 2
print("--- Tienda 2 ---")
print(tienda2['Categoría del Producto'].value_counts().head())
print("-" * 35)

# Tienda 3
print("--- Tienda 3 ---")
print(tienda3['Categoría del Producto'].value_counts().head())
print("-" * 35)

# Tienda 4 (Menor Facturación)
print("--- Tienda 4 (Menor Facturación) ---")
print(tienda4['Categoría del Producto'].value_counts().head())
print("-" * 35)
```

## Gráfico de Barras Agrupado/Líneas: Rendimiento de Métricas Clave

```python
import numpy as np

# Datos
tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
calificaciones = [3.98, 4.04, 4.05, 4.00]
facturacion_millones = [f / 1000000 for f in facturacion] # Escalar para el gráfico

fig, ax1 = plt.subplots(figsize=(10, 6))

# Gráfico de Barras (Eje Y Izquierda) para Facturación
ax1.bar(tiendas, facturacion_millones, color='gray', alpha=0.6, label='Facturación (Millones)')
ax1.set_xlabel('Tienda')
ax1.set_ylabel('Facturación Total (Millones)', color='gray')
ax1.tick_params(axis='y', labelcolor='gray')

# Crear el segundo eje Y para Calificación
ax2 = ax1.twinx()
ax2.plot(tiendas, calificaciones, color='blue', marker='o', linestyle='-', linewidth=2, label='Calificación Promedio')
ax2.set_ylabel('Calificación Promedio (1 a 5)', color='blue')
ax2.tick_params(axis='y', labelcolor='blue')
ax2.set_ylim(3.8, 4.2) # Ajustar para resaltar pequeñas diferencias

plt.title('Facturación vs. Calificación Promedio por Tienda')
fig.tight_layout()
plt.show()
```

Las cuatro tiendas muestran un patrón de ventas extremadamente similar en cuanto a la popularidad de las categorías: **Muebles, Electrónicos y Juguetes son consistentemente las más vendidas.**

Lo interesante es que la Tienda 4 (la de menor facturación) tiene un volumen de ventas en su Top 5 categorías casi idéntico al de la Tienda 1 (la de mayor facturación), e incluso tiene el mayor volumen de ventas de Juguetes.

**Conclusión:** El problema de la Tienda 4 no es la demanda por categoría (la gente compra los productos populares). La menor facturación total **($1.038 mil millones)** debe deberse a:

Precio promedio más bajo por transacción (vende artículos más baratos).

Menos ventas totales en categorías que no son del Top 5.

Esto refuerza la candidatura de la Tienda 4 como la menos eficiente.

# 3. Calificación promedio de la tienda


```python
# Analizar la Calificación Promedio de los Clientes
print("Clasificación Promedio de los Clientes por Tienda \n")

# Calcular el promedio de las calificaciones
promedio_t1 = tienda['Calificación'].mean()
promedio_t2 = tienda2['Calificación'].mean()
promedio_t3 = tienda3['Calificación'].mean()
promedio_t4 = tienda4['Calificación'].mean()

print(f"Calificación Promedio Tienda 1: {promedio_t1:.2f}")
print(f"Calificación Promedio Tienda 2: {promedio_t2:.2f}")
print(f"Calificación Promedio Tienda 3: {promedio_t3:.2f}")
print(f"Calificación Promedio Tienda 4: {promedio_t4:.2f}")

```

Las calificaciones promedio de los clientes son uniformemente altas en todas las tiendas (entre 3.98 y 4.05). De hecho, la Tienda 4 tiene una calificación de 4.00, ligeramente superior a la Tienda 1 (3.98).

**Conclusión:** La baja facturación de la Tienda 4 no se debe a un problema de calidad del producto ni de servicio al cliente. Los clientes están satisfechos, lo que sugiere que la ineficiencia es de carácter logístico o de precios.

##Importar Librerías y Plugins
Asegúrate de tener instalados e importados Folium y su plugin HeatMap.

```python
import pandas as pd
import folium
from folium.plugins import HeatMap
```

##Preparación de Datos de Calor
Usaremos solo los datos de la Tienda 4 (tienda4) para mapear su concentración de ventas.

```python
# Asumiendo que tu DataFrame 'tienda4' ya está cargado.
# Preparamos los datos en el formato [lat, lon] que HeatMap necesita.
datos_calor_t4 = tienda4[['lat', 'lon']].values.tolist()

# Calculamos el punto central de la Tienda 4 para centrar el mapa
latitud_central_t4 = tienda4['lat'].mean()
longitud_central_t4 = tienda4['lon'].mean()
```

##Crear el Mapa Base y Agregar HeatMap
Creamos el mapa y agregamos la capa de calor con los datos de la Tienda 4.

```python
# 1. Crear el mapa base centrado
mapa_calor_t4 = folium.Map(
    location=[latitud_central_t4, longitud_central_t4],
    zoom_start=5
)

# 2. Agregar la capa de Mapa de Calor (HeatMap)
HeatMap(datos_calor_t4).add_to(mapa_calor_t4)
```

Simplemente llama al objeto del mapa al final de la celda de código:

```python
# Mostrar el mapa interactivo en Colab
mapa_calor_t4
```

##Conclusión y Recomendación Final



El Mapa de Calor de las ventas de la Tienda 4, en conjunto con el análisis de la baja facturación, ofrece un insight geográfico clave:

**Problema de Densidad:** El mapa muestra que la actividad de ventas de la Tienda 4 está fuertemente concentrada en una única y limitada región (el eje central del país, cerca de Ibagué/Meta).

**Falla en la Penetración**: La Tienda 4 falla en generar una densidad de ventas comparable en otras áreas clave y de alta actividad comercial del país (como la costa), donde sus competidores sí son fuertes.

Esta visualización confirma que la ineficiencia de la Tienda 4 no es un problema de mala ubicación o de una logística costosa, sino un problema de baja penetración de mercado fuera de su nicho principal. Esto refuerza la decisión de que la Tienda 4 es el activo de menor rendimiento y el candidato lógico para la venta.
